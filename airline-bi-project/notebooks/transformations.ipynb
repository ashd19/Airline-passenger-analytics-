```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("customer_segments.csv")
print(df.head(5))
df.columns
```

```python
df['Satisfaction_Binary'] = df['Satisfaction'].map({
    'Satisfied': 1,
    'Neutral or Dissatisfied': 0
})
df['Total Delay'] = df['Departure Delay'] + df['Arrival Delay']
def delay_category(x):
    if x == 0:
        return "On Time"
    elif x <= 15:
        return "Minor Delay"
    elif x <= 60:
        return "Moderate Delay"
    else:
        return "Severe Delay"

df['Delay Category'] = df['Total Delay'].apply(delay_category)
df['High Delay'] = (df['Total Delay'] > 60).astype(int)
df['Delayed'] = (df['Total Delay'] > 0).astype(int)
df.columns
df.dtypes
```

```python
df.to_csv("final_data.csv", index=False)
```
